## 牛啊
- 直接用快慢指针，慢的在后面追，
- 如果是有环的话
- 最差也是一整个环，也会在最后相遇
- 如果是没有环，那么慢指针会走到最后，最差也还是O(N)


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        f = head
        s = head

        while f and f.next:
            f = f.next.next
            s = s.next
            if s == f:
                return True
        return False
        

        
```

证明过程
- 环的存在性：

假设链表中存在环，环外部分长度为a，环的周长为b。
 slow指针进入环后，每走b步，fast指针会走2b步，因此fast相对于slow多走b步，即每走b步fast就会追上slow一次。


- 环的入口：

当slow和fast在环内相遇时，设相遇点为M。
从链表头部到相遇点M的距离为x（slow走过的距离），从环入口到相遇点M的距离为y（fast走过的距离减去环的周长）。
由于fast的速度是slow的两倍，所以fast走过的距离是slow的两倍，即2x = x + nb，其中n是fast在环内走过的圈数。
因此，x = nb，即从链表头部到相遇点M的距离等于环的周长乘以某个整数。


- 找到环的入口：

当slow和fast在环内相遇后，将slow重新指向链表头部，fast保持在相遇点M。
然后，slow和fast都以每次一步的速度移动，直到它们再次相遇。
由于slow和fast都以相同的速度移动，且它们之间的距离是环的周长的整数倍，所以它们会在环的入口处相遇。
