## 先走一遍将新节点与旧节点对应起来，并创建新节点
## 然后将旧节点东西全复制过去

```python
class Solution:
    def copyRandomList(self, head: 'Optional[Node]') -> 'Optional[Node]':
        if not head:
            return None
        cur = head
        node_dict = {}
        while cur:
            node_dict[cur] = Node(cur.val)  # 将旧节点与新节点对应起来（应该是内存的地方）
            cur = cur.next
        cur = head
        while cur:
            if cur.next:# 要将旧的节点对应的新的下一个节点从哈希表中找出来
                node_dict[cur].next = node_dict[cur.next]
            if cur.random:
                node_dict[cur].random = node_dict[cur.random]
            cur = cur.next

        return node_dict[head]
```
