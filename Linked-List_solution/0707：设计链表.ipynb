```python
#
# @lc app=leetcode.cn id=707 lang=python3
#
# [707] 
#

# @lc code=start
class Linked_Node():
    def __init__(self, data = 0, next = None):
        self.next = next
        self.data = data
        

class MyLinkedList:

    def __init__(self):
        self.head = Linked_Node(0)
        self.size = 0
        self.flag = True

    def get(self, index: int) -> int:
        if self.flag == True and index == 0:
            return -1  #  
        count = 0
        if self.size < index:
            return -1 
        cur = self.head
        while count < index:
            cur = cur.next
            count += 1
        return cur.data

    def addAtHead(self, val: int) -> None:
        if not self.flag:  # 
            self.size += 1
            node = Linked_Node(val)
            node.next = self.head
            self.head = node 
            
        else:
            self.head.data = val
            self.flag = False


    def addAtTail(self, val: int) -> None:
        if self.flag == True:  # 
            return self.addAtHead(val)
        count = 0
        cur = self.head
        while count < self.size:
            count += 1
            cur = cur.next
        node = Linked_Node(val)
        cur.next = node
        self.size += 1

    def addAtIndex(self, index: int, val: int) -> None:
        if index == self.size + 1 and self.flag == False :  # size + 1size 
            self.addAtTail(val)
        elif index == 0:
            self.addAtHead(val)
        elif index > self.size + 1:
            return      
        else:
            count = 0
            cur = self.head
            while count < index - 1:
                cur = cur.next
                count += 1
            node = Linked_Node(val)
            node.next = cur.next
            cur.next = node
            self.size += 1
            

    def deleteAtIndex(self, index: int) -> None:
        if index == 0:
            self.head = self.head.next
            self.size -= 1
        elif index > self.size:
            return None
        else:
            count = 0
            cur = self.head
            while count < index - 1:
                count += 1
                cur = cur.next
            cur.next = cur.next.next
            self.size -= 1
        

obj = MyLinkedList()
obj.addAtIndex(1, 0)
param_1 = obj.get(0)
'''obj.deleteAtIndex(2) 
obj.addAtHead(6)
obj.addAtTail(4)

obj.addAtHead(4)
obj.addAtIndex(5, 0)
obj.addAtHead(6)'''
#obj.deleteAtIndex(1) 
#param_2 = obj.get(1)
# obj.addAtHead(val)
# obj.addAtTail(val)
count = 0
cur = obj.head

while cur:
    print(cur.data, end = " -> ")
    cur = cur.next
print("None")
print(obj.size)
print(param_1)
```
