```python
from typing import Optional

class ListNode:
    def __init__(self, val = 0, next = None):
        self.val = val
        self.next = next
class Solution:
    def isPalindrome(self, head: Optional[ListNode])-> Optional[ListNode]:
        # 
        cur = head
        test = []
        while cur:
            test.append(cur.val)
            cur = cur.next
        return test[::1] == test[::-1]



array = [1, 2, 2, 1]
head = ListNode(array[0])
cur = head
for i in range(1, len(array)):
    node = ListNode(array[i])
    cur.next = node
    cur = cur.next

test = Solution()
head = test.isPalindrome(head)
head        
```
