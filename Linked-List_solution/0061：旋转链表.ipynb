```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

from typing import Optional
class Solution:
    def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        if not head:
            return head

        # size - k 
        cur = head
        size = 1
        while cur.next :
            cur = cur.next
            size += 1
        cur.next = head
        cur = head
        # size > k !!!!!!!
        if size < k:
            k = k % size
        elif size == k:
            return head
    
        for _ in range(size - k - 1):#
            cur = cur.next
        temp = cur.next # 
        cur.next = None
        return temp

array = [1]
head = ListNode(array[0])
cur = head
for i in range(1, len(array)):
    node = ListNode(array[i])
    cur.next = node
    cur = cur.next

test = Solution()
head = test.rotateRight(head, 1)


```
