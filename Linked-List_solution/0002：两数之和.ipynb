```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        result = ListNode()
        cur = result
        carry = 0 # 进位
        cur1 = l1
        cur2 = l2
        while cur1 and cur2:
            value = (cur1.val + cur2.val + carry) % 10
            carry = (cur1.val + cur2.val + carry) // 10
            node = ListNode(value)
            cur.next = node
            cur = cur.next
            cur1 = cur1.next
            cur2 = cur2.next
        
        if cur1:
            while cur1:
                value = (cur1.val + carry) % 10
                carry = (cur1.val + carry) // 10
                node = ListNode(value)
                cur.next = node
                cur = cur.next 
                cur1 = cur1.next
            if carry != 0:
                cur.next = ListNode(1, None)
        elif cur2:
            while cur2:
                value = (cur2.val + carry) % 10
                carry = (cur2.val + carry) // 10
                node = ListNode(value)
                cur.next = node
                cur = cur.next 
                cur2 = cur2.next    
            if carry != 0:
                cur.next = ListNode(1, None)
        elif carry != 0:
            cur.next = ListNode(1)
        return result.next
```
