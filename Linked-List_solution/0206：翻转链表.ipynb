```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        pre = None
        cur = head
        while cur:
            temp = cur.next
            cur.next = pre
            pre = cur
            cur = temp
        return pre

# 



# debug 
array = [1, 2, 3, 4, 5]
head = ListNode(array[0])
cur = head
for i in range(1, len(array)):
    node = ListNode(array[i])
    cur.next = node
    cur = cur.next

test = Solution()
head = test.reverseList(head)

cur = head
while cur:
    print(cur.val, end = "->")
    cur = cur.next
print("None")
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def reverse(self, pre, cur):
        if cur == None:
            return pre
        temp = cur.next
        cur.next = pre
        return self.reverse(cur, temp)
        
    def reverseList(self, head: ListNode) -> ListNode:
        return self.reverse(None, head)

# 



# debug 
array = [1, 2, 3, 4, 5]
head = ListNode(array[0])
cur = head
for i in range(1, len(array)):
    node = ListNode(array[i])
    cur.next = node
    cur = cur.next

test = Solution()
head = test.reverseList(head)

cur = head
while cur:
    print(cur.val, end = "->")
    cur = cur.next
print("None")






```
