```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def middleNode(self, head: Optional[ListNode]) -> Optional[ListNode]:
        fast = slow = head
        # while fast == None or fast.next == None:
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        return slow
```
