## 有序数组且存在一个有效答案
一个遍历一个二分？


```python
# 一个遍历一个二分试试
from typing import List


class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        size = len(numbers)
        for i in range(size):
            left = 0
            right = size - 1
            if numbers[i] >= target:
                right = i - 1
            elif numbers[i] < target:
                left = i + 1
            while left <= right:
                mid = (left + right) // 2
                if numbers[mid] < target - numbers[i]:
                    left = mid + 1
                elif numbers[mid] > target - numbers[i]:
                    right = mid - 1
                else:
                    return sorted([mid + 1, i + 1])    # 他的下标从1开始

a = Solution()
print(a.twoSum([2,3,4],6))
            
```

从两边开始
如果大了就大的一端左移
如果小了就小的一端右移动

```python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        l, r = 0, len(numbers)-1
        while l < r:
            if numbers[l] + numbers[r] > target:
                r -= 1
            elif numbers[l] + numbers[r] < target:
                l += 1
            else:
                return [l+1, r+1]
```
