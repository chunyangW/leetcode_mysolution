
128.   最长连续序列
中等
给定一个未排序的整数数组 nums ，找出数字连续的最长序列（不要求序列元素在原数组中连续）的长度。

请你设计并实现时间复杂度为 O(n) 的算法解决此问题。

# 原来存放到set中是O（n）的，查找也是O（n）的(n个O（1）)


```python
from typing import List

class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        s = set(nums)
        res = 0
        for num in s:
            if num - 1 not in s:
                curr_num = num
                curr_streak = 1
                
                while curr_num + 1 in s:
                    curr_num += 1
                    curr_streak += 1
                res = max(res, curr_streak)
        return res
                
```
