给你一个整数数组 nums 。如果任一值在数组中出现 至少两次 ，返回 true ；如果数组中每个元素互不相同，返回 false 。

````python
from typing import List
# 使用集合
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        table = set()
        for item in nums:
            if item in table: # 提前结束
                return True
            table.add(item)
        return False
````

在Python中，`set` 是一个内置的数据类型，它是一个无序的集合，可以存储不重复的元素。以下是一些基本的 `set` 函数用法：

1. **创建集合**：
   ```python
   my_set = set()
   my_set = {1, 2, 3}
   my_set = {x for x in range(5)}  # 使用集合推导式
   ```

2. **添加元素**：
   ```python
   my_set.add(4)
   ```

3. **删除元素**：
   ```python
   my_set.remove(2)  # 删除元素2，如果元素不存在会抛出KeyError
   my_set.discard(3)  # 删除元素3，如果元素不存在不会抛出错误
   my_set.pop()       # 随机删除一个元素
   ```

4. **清空集合**：
   ```python
   my_set.clear()  # 清空集合
   ```

5. **检查元素是否存在**：
   ```python
   if 1 in my_set:
       print("1 is in the set")
   ```

6. **集合的大小**：
   ```python
   size = len(my_set)
   ```

7. **集合的交集**：
   ```python
   a = {1, 2, 3}
   b = {2, 3, 4}
   intersection = a & b  # 或者 a.intersection(b)
   ```

8. **集合的并集**：
   ```python
   a = {1, 2, 3}
   b = {2, 3, 4}
   union = a | b  # 或者 a.union(b)
   ```

9. **集合的差集**：
   ```python
   a = {1, 2, 3}
   b = {2, 3, 4}
   difference = a - b  # 或者 a.difference(b)
   ```

10. **集合的对称差集**：
    ```python
    a = {1, 2, 3}
    b = {2, 3, 4}
    symmetric_difference = a ^ b  # 或者 a.symmetric_difference(b)
    ```

11. **将集合转换为列表或元组**：
    ```python
    list_set = list(my_set)
    tuple_set = tuple(my_set)
    ```

12. **集合的迭代**：
    ```python
    for element in my_set:
        print(element)
    ```

13. **集合的复制**：
    ```python
    new_set = set(my_set)
    ```

14. **集合推导式**：
    ```python
    squares = {x**2 for x in range(10)}
    ```

集合是一个强大的数据结构，特别适合于需要快速成员检查、元素唯一性以及集合运算的场景。
