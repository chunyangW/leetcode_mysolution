```python
from typing import List
class Solution:
    def numSubarrayProductLessThanK(self, nums: List[int], k: int) -> int:
        # 右边边界一直向右边移动
        # 应该是维护一个总乘积小于k的窗口
        # 然后收缩
        # 然后加上 右边元素与每一个窗口元素构成的子数组 right - left + 1
        temp = 1
        res = 0
        left = 0
        if k <= 1:
            return 0
        for right, val in enumerate(nums):
            temp *= val
            while temp >= k:
                temp = temp / nums[left]
                left += 1
            res += (right - left + 1)
        return res
```
