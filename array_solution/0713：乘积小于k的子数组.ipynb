```python
from typing import List
class Solution:
    def numSubarrayProductLessThanK(self, nums: List[int], k: int) -> int:
        # 
        # k
        # 
        #   right - left + 1
        temp = 1
        res = 0
        left = 0
        if k <= 1:
            return 0
        for right, val in enumerate(nums):
            temp *= val
            while temp >= k:
                temp = temp / nums[left]
                left += 1
            res += (right - left + 1)
        return res
```
