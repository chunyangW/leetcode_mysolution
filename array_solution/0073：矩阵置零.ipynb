```python
from typing import List

# 思路：如果先初始化一个Mask 矩阵，如果该行列有0那么就使得mask为0
# 然后将mask 与原矩阵相乘得到最终结果
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        i = len(matrix)
        j = len(matrix[0])
        mask =[[ 1 for m in range(j)] for n in range(i)]     # 先行后列 才是对的
        for m in range(i):
            for n in range(j):
                if matrix[m][n] == 0:  # 将掩码矩阵第m行与第n列设置为0
                    for col in range(j):
                        mask[m][col] = 0
                    for row in range(i):
                        mask[row][n] = 0
        for m in range(i):
            for n in range(j):
                matrix[m][n] = matrix[m][n] * mask[m][n]
        print(mask)


mat = [[0,1,2,0],[3,4,5,2],[1,3,1,5]]
a = Solution()
a.setZeroes(mat)
print (mat)
```

```python
# 也可以直接记录哪一列    需要  变为0
# 那么就是说mask 和 这个方法都是再记录哪里需要变为0
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        m = len(matrix)
        n = len(matrix[0])
        
        # 初始化两个列表，用于记录哪些行和列需要设置为0
        lines = [0] * m
        rows = [0] * n
        
        # 第一遍遍历矩阵，记录哪些行和列需要设置为0
        for i in range(m):
            for j in range(n):
                if matrix[i][j] == 0:
                    lines[i] = 1  # 第i行需要设置为0
                    rows[j] = 1  # 第j列需要设置为0
        
        # 第二遍遍历矩阵，根据记录的行和列信息，将相应的元素设置为0
        for i in range(m):
            for j in range(n):
                if lines[i] or rows[j]:
                    matrix[i][j] = 0
```
