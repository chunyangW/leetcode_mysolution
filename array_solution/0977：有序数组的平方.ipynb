```python
from typing import List

class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        size = len(nums)
        left, right = 0, size - 1
        res = []
        while left < right:
            l, r = abs(nums[left]), abs(nums[right])
            if l < r:
                res.append(r**2)
                right -= 1
            elif l > r:
                res.append(l**2)
                left += 1
            else:
                res.append(r**2)
                res.append(r**2)
                left += 1
                right -= 1
        if left == right:  # 
            res.append(nums[left]**2)
        return res[-1 : : -1]
    
a = Solution()
a.sortedSquares([-10000,-9999,-7,-5,0,0,10000])
```
