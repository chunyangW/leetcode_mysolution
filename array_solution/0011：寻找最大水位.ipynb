## 由于每次水量的多少只取决于最小的水位与两者之间的距离
所以每次移动较小的一端的指针就好


```python
from typing import List

class Solution:

    def maxArea(self, height: List[int]) -> int:
        size = len(height)
        l, r, ans = 0, size - 1, 0
        while l <= r:
            ans = max((r - l) * min(height[l], height[r]), ans)
            if height[l] < height[r]:
                l += 1
            else:
                r -= 1
        return ans
    
a = Solution()
a.maxArea([1,8,6,2,5,4,8,3,7])

```
