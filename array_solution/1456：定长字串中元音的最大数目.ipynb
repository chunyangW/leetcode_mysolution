```python
from typing import List
from math import inf

class Solution:
    def maxVowels(self, s: str, k: int) -> int:
        # 还是固定长度窗口求最多元素类问题
        size = len(s)
        left, right, temp, res = 0, 0, 0, -inf
        table = ['a', 'e', 'i', 'o', 'u']
        while right < size and left <= right:
            while right - left <= k - 1:
                if s[right] in table:
                    temp += 1
                right += 1
            res = max(res, temp)
            if s[left] in table:
                temp -=  1
            left += 1 
        return res

```
