```python
from typing import List

class Solution:
    def longestOnes(self, nums: List[int], k: int) -> int:
        # 0k
        # 
        left, right, ans = 0, 0, 0
        count = 0
        size = len(nums)
        while right < size:
            if nums[right] == 0:
                count += 1
                if count > k:
                    ans = max(ans, right - left)
                    while count > k:
                        if nums[left] == 0:      
                            count -= 1
                        left += 1
            right += 1
        # count:
        if count <= k:
            return max(ans, right - left)
        return ans
```
