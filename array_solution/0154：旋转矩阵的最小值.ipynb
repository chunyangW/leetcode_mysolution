```python
from typing import List

class Solution:
    def findMin(self, nums: List[int]) -> int:
        l, r = 0, len(nums) - 1
        while l < r:
            mid = (l + r) // 2
            if nums[mid] > nums[r]:  # 
                l = mid +  1
            elif nums[mid] < nums[l]: # mid
                r = mid
            else:                      # mid
                r -= 1             # rr--
        return nums[l]

a = Solution()
a.findMin([1,3,5])
```
