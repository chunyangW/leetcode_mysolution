```python
from typing import List

class Solution:
    def findMin(self, nums: List[int]) -> int:
        l, r = 0, len(nums) - 1
        while l < r:
            mid = (l + r) // 2
            if nums[mid] > nums[r]:  # 右边必定出现最小值
                l = mid +  1
            elif nums[mid] < nums[l]: # 左边必定出现最小值且mid可能是最小的
                r = mid
            else:                      # 如果出现相等的那么调整mid
                r -= 1             # 为什么要调整r呢因为如果上面条件都不满足，说明最后是有序的了，那么就取最小的也就是让r--就好
        return nums[l]

a = Solution()
a.findMin([1,3,5])
```
