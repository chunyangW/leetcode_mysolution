- 不同的地方在于要统计长度了然后取最小长度来添加

```python
from typing import List

class Solution:
    def intersect(self, nums1: List[int], nums2: List[int]) -> List[int]:
        intersect = []
        nums1.sort()
        nums2.sort()
        len1, len2 = len(nums1), len(nums2)
        index1, index2, temp1, temp2 = 0, 0, 1, 1

        def Append(num, nums, count):
            for _ in range(count):
                nums.append(num)
        
        while index1 < len1 and index2 < len2:
            num1 = nums1[index1]
            num2 = nums2[index2]
            if num1 == num2:
                index1 += 1
                index2 += 1
                while index2 < len2 and nums2[index2] == nums2[index2 - 1]:
                    index2 += 1
                    temp2 += 1 
                while index1 < len1 and nums1[index1] == nums1[index1 - 1]:
                    index1 += 1
                    temp1 += 1
                Append(num1, intersect, min(temp1, temp2))
                temp1 = temp2 = 1
            elif num1 > num2:
                index2 += 1
            else:
                index1 +=1
        return intersect
    
a = Solution()
a.intersect([1,3,8,9,3],[1,0])


                
```
