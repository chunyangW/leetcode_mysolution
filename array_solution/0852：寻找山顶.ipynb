# 二分查找
- 中间小于左边，峰值在左边
- 中间小于右边，峰值在右边

```python
from typing import List

class Solution:
    def peakIndexInMountainArray(self, arr: List[int]) -> int:
        l, r = 0, len(arr)-1
        while l <= r:
            mid = (l + r) // 2
            if arr[mid] < arr[mid - 1]:
                r = mid 
            elif arr[mid] < arr[mid + 1]:
                l = mid 
            else: 
                return mid
        return l
            
a = Solution()
a.peakIndexInMountainArray([3,9,8,6,4])
```
