## left < mid < right 

```python
#
from typing import List

class Solution:
    def findMin(self, nums: List[int]) -> int:
        left, right = 0, len(nums) - 1
        while left < right:
            mid = (left + right) // 2
            if nums[mid] > nums[right]:  # 
                left = mid + 1
            else:     # 
                right = mid  #midmid
        return left


a = Solution()
a.findMin([3,1,2])
    


```
