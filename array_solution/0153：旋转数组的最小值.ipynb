## 因为是单调递增的，但是只有在最小值处不是递增
# 一小段是无序的出现最小值，但是什么时候无序？
## left < mid < right 就是有序

```python
#
from typing import List

class Solution:
    def findMin(self, nums: List[int]) -> int:
        left, right = 0, len(nums) - 1
        while left < right:
            mid = (left + right) // 2
            if nums[mid] > nums[right]:  # 右边无序了出现了最小值
                left = mid + 1
            else:     # 左边无序
                right = mid  #mid有可能直接是最小值，所以要取mid
        return left


a = Solution()
a.findMin([3,1,2])
    


```
