### 不是查找最小数值了
好像也还是查找最小数值
查到最小数值后分成两半


```python
from typing import List

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        l, r = 0, len(nums) - 1
        while l < r:
            mid = (l + r) // 2
            if nums[mid] > nums[r]:  # 右边必定出现最小值
                l = mid +  1
            elif nums[mid] < nums[l]: # 左边必定出现最小值且mid可能是最小的
                r = mid
            else:                      # 如果出现相等的那么调整mid
                r -= 1             # 为什么要调整r呢因为如果上面条件都不满足，说明最后是有序的了，那么就取最小的也就是让r--就好
        # l 就是最小值了
        mi = l 
        
        # 确定有序列范围
        l, r = 0, len(nums) - 1
        if nums[0] <= target <= nums[mi]:
            r = mi
        else:
            l = mi
        while l < r:
            mid = (l + r) // 2
            if nums[mid] < target :
                l = mid + 1
            elif nums[mid] > target:
                r = mid - 1
            else:
                return mid
        return l
```
