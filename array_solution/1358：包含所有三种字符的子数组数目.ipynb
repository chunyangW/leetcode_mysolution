```python
from typing import List
from collections import defaultdict

class Solution:
    def numberOfSubstrings(self, s: str) -> int:
        # 
        # 
        #  l  r  [l,r]  abc 
        #  l  r  len(s)r 
        #  l  s[l]  [l+1,r]  abc 
        #  r 

        size = len(s)
        left = 0
        res = 0
        cnt = defaultdict(int)
        for right, val in enumerate(s):
            cnt[val] += 1
            while len(cnt) == 3:
                res += size - right
                if cnt[s[left]] == 1:
                    del cnt[s[left]]
                else:
                    cnt[s[left]] -= 1
                left += 1
        return res 

a = Solution()
print(a.numberOfSubstrings("aaaabc"))
```
