```python
from typing import List

class Solution:
    def numOfSubarrays(self, arr: list[int], k: int, threshold: int) -> int:
        left, right = 0, 0
        window = 0
        count = 0

        while right < len(arr):  
            window += arr[right]  #  
            # right = left = 0windowarr[0]
            if  right - left + 1 >= k: # k  left++
                if window >= k * threshold:
                    count += 1
                window -= arr[left]
                left += 1
            right += 1
        return count


a = Solution()
a.numOfSubarrays([2,2,2,2,5,5,5,8],3,4)
```
