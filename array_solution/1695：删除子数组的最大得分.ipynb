```python
from typing import List

class Solution:
    def maximumUniqueSubarray(self, nums: List[int]) -> int:
        # 1 
        # val1
        # sum
        size = len(nums)
        c = {}
        res, temp, right, left = 0, 0, 0, 0
        while right < size:
            c[nums[right]] = c.get(nums[right], 0) + 1  #  0
            temp += nums[right]

            while c[nums[right]] > 1: # 
                temp -= nums[left]
                c[nums[left]] -= 1
                left += 1

            res = max(res, temp)  # 
            right += 1
        return res
```
