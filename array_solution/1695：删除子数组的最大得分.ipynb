```python
from typing import List

class Solution:
    def maximumUniqueSubarray(self, nums: List[int]) -> int:
        # 维护一个哈希表使得窗口中元素数目小于等于1 
        # 每次右移动一个，然后使得左指针向右移动直到右指针的val为1
        # 然后每一次记录sum
        size = len(nums)
        c = {}
        res, temp, right, left = 0, 0, 0, 0
        while right < size:
            c[nums[right]] = c.get(nums[right], 0) + 1  # 扩张 没有就赋初值为0
            temp += nums[right]

            while c[nums[right]] > 1: # 缩短
                temp -= nums[left]
                c[nums[left]] -= 1
                left += 1

            res = max(res, temp)  # 记录每次的最大值
            right += 1
        return res
```
