#  $i = 2 $ 
# 
- 
- left + 1
- left  right - 1 right - 1

```python
from typing import List

class Solution:
    def triangleNumber(self, nums: List[int]) -> int:
        nums = sorted(nums)
        size = len(nums)
        i, ans = 2, 0
        while i < size:
            left, right = 0, i - 1
            while left < right:
                if nums[left] + nums[right] <= nums[i]:
                    left += 1
                else:
                    ans += (right - left)
                    right -= 1
            i += 1
        return ans
a = Solution()
a.triangleNumber([4,2,3,4])
            
```
