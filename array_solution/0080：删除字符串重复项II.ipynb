```python
from typing import List

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        count, slow, fast = 1, 0, 0
        size = len(nums)
        if size == 1:
            return 1
        
        while fast < size - 1:
            fast += 1
            if nums[fast] == nums[fast - 1]:
                count += 1   # count ....
                # 
                if count == 2:
                    while fast < size and nums[fast] == nums[fast - 1]:
                        fast += 1

                    nums[slow], nums[slow + 1]= nums[fast - 1], nums[fast - 1]
                    slow += 2
                count = 1
                continue
            # 
            else:
                count = 1
                nums[slow] = nums[fast - 1]
                slow += 1
            
        # 
        if nums[fast] != nums[fast - 1]:
                nums[slow] = nums[fast]
                slow += 1 
        print(nums[:slow])
        return slow

a = Solution()
a.removeDuplicates([0,0,1,1,1,1,2,3,3])
```
