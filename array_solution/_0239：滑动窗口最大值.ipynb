```python
from typing import List
from collections import deque
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        # i >= k + q[0] pop
        # 
        #         
        # 
        # 
        #  
        # 
        # 
        #      
        q = deque()
        N = len(nums)
        res = []
        for i, x in enumerate(nums):
            # ,
            while q and q[-1] <= x:
                q.pop()
            q.append(i)
            # 
            if q[0] + k <= i:
                q.popleft()
            # 
            if i >= k - 1:
                res.append(nums[q[0]])
        return res
    
a = Solution()
a.maxSlidingWindow([7,2,4], 2)
```
