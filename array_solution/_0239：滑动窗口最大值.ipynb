```python
from typing import List
from collections import deque
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        # 记下最大值出现的索引，如果当前索引i >= k + q[0] 就pop
        # 还是使用队列来登记最大值的索引的位置
        # 不足的时候先找初始窗口中的最大值        
        # 但是有个问题啊，当最大的弹出来了，那最大到当前之前的次最大在哪，我怎么确定
        # 使用一个队列就好了来一个最大记录一下当前索引放在最前面，然后每次遍历就比较当前遍历的元素与队列最前面元素的大小
        # 同时每次到一个最大的地方就判断一下要不要将最后一个元素弹出因为 在前面的队列元素肯定是后面才来的所以一定是队列最后的元素才需要弹出
        # 所以我需要再记录次最大的数字并且每次更新
        # 错我需要维护一个单调队列同时每次插入新的元素所处的位置
        # 来了一个新的，只要比他小的前面的老东西就全部拿走，反正后面一定用不上了     
        q = deque()
        N = len(nums)
        res = []
        for i, x in enumerate(nums):
            # 新来一个将前面的老东西杀完,相等的也要杀掉
            while q and q[-1] <= x:
                q.pop()
            q.append(i)
            # 还有自然死亡的老东西
            if q[0] + k <= i:
                q.popleft()
            # 添加
            if i >= k - 1:
                res.append(nums[q[0]])
        return res
    
a = Solution()
a.maxSlidingWindow([7,2,4], 2)
```

```python
# deque 双端队列可以在头尾插入也能在中间某一处插入
# 在队列的开头添加元素
deque.appendleft('x')

# 在队列的任意位置插入元素
deque.insert(2, 'y')  # 在索引2的位置插入元素'y'

# 从队列的开头移除元素
deque.popleft()

# 从队列的末尾移除元素
deque.pop()

# 从队列的任意位置移除元素
deque.pop(2)  # 移除索引2的元素
```
