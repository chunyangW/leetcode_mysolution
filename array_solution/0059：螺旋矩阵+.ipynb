```python
from typing import List
import numpy as np

class Solution:
    def generateMatrix(self, n: int) -> List[List[int]]:
        rows = cols = n
        all = int(rows/2 + 0.5)              # 循环次数
        start = -1   # 下面是循环每次更迭起始索引为 0，0 -》1，1 -》2，2.。。
        res = [[1 for _ in range(cols)] for _ in range(rows) ]   # 初始化结果矩阵
        val = 1  #从一开始赋值

        for k in range(all):
            start += 1
            i = j = start    # 更新索引到新的起始点
            while ( j < cols - k):  # 第一行
                res[i][j] = val
                val += 1
                j += 1
                
            j -= 1
            i += 1

            while(  i < rows - k): # 最后一列  上面的j = cols - k 所以需要将j - 1 还有一个问题是里面不够一整个循环了 加个len(res) 判断就好了
                res[i][j] = val
                val += 1
                i += 1 
            i -= 1            
            j -=1 

            while( j >= k and rows!= 0):# 最后一行  会有 [[2,3]]
                res[i][j] = val
                val += 1
                j -= 1
            j += 1
            i -=1

            while(i >= k+1 and cols != 1):#第一列
                res[i][j] = val
                val += 1
                i -=1
        return res

a = Solution()
a.generateMatrix(10)
```
