```python
from typing import List

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums = sorted(nums)
        size = len(nums)
        ans = []
        i = size - 1 
        while i < size:
            left, right = 0, i - 1
            if i != size - 1  and nums[i] == nums[i + 1]:
                i -= 1
                continue
            while left < right:
                if nums[left] + nums[right] < (-nums[i]):
                    left += 1
                elif nums[left] + nums[right] > (-nums[i]):
                    right -= 1
                else:
                    ans.append([nums[left], nums[right], nums[i]])
                    # left right right --
                    right -= 1
                    left += 1
                    # leftright-1left
            i -= 1
        return ans
    
a = Solution()
a.threeSum([-2,0,1,1,2])

```
