## 和611有效三角形个数问题类似试一下
## 第三个数不能从第二个开始

```python
from typing import List

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums = sorted(nums)
        size = len(nums)
        ans = []
        i = size - 1 
        while i < size:
            left, right = 0, i - 1
            if i != size - 1  and nums[i] == nums[i + 1]:
                i -= 1
                continue
            while left < right:
                if nums[left] + nums[right] < (-nums[i]):
                    left += 1
                elif nums[left] + nums[right] > (-nums[i]):
                    right -= 1
                else:
                    ans.append([nums[left], nums[right], nums[i]])
                    # 因为在这个时候是相等了，对的那么left 到right 之间都不行了，那么需要right --
                    right -= 1
                    left += 1
                    # 同时left变大，right-1了left肯定不能变小了
            i -= 1
        return ans
    
a = Solution()
a.threeSum([-2,0,1,1,2])

```
