```python
from typing import List

class Solution:
    def search(self, nums: List[int], target: int) -> int:

        # 先初始化左右边界
        left, right = 0, len(nums) - 1

        # 再循环找
        while left < right:

            # 二分查找中间值
            mid = (left + right) // 2
            if target > nums[mid]:
                left = mid + 1
            # 下面一定是 elif不然 if 如果没通过 下面的else直接通过了
            elif target < nums[mid]:        
                right = mid - 1
            else:
                return mid
            
        # 啥也没找到
        return left if nums[left] == target else -1 # 与上面的 while left < right： 相关 中止的时候一定是left == right 的这是一个好处

a= Solution()
a.search([-1,0,3,5,9,12], 9)
```
