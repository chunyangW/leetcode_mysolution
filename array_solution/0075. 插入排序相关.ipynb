### 插入排序示例

````python
class Solution:
    def insertionSort(self, nums: [int]) -> [int]:
        # 遍历无序区间
        for i in range(1, len(nums)):
            temp = nums[i]
            j = i
            # 从右至左遍历有序区间
            while j > 0 and nums[j - 1] > temp:
                # 将有序区间中插入位置右侧的元素依次右移一位
                nums[j] = nums[j - 1]
                j -= 1
            # 将该元素插入到适当位置
            nums[j] = temp

        return nums

    def sortArray(self, nums: [int]) -> [int]:
        return self.insertionSort(nums)
````

# 插入排序解法：

````python
from typing import List


class Solution:
    def sortColors(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        self.insertsort(nums)


    def insertsort(self, nums: List[int]) -> None:
        size = len(nums)
        for i in range(1,size):
            # 第i个（第j个）是我们未排序的第一个
            # 下面要遍历的从j-1开始都是我们排序好的数列部分
            j = i 
            temp = nums[i]
            while(j > 0 and nums[j-1] > temp):
                # 将大于待插入数字的向右移动
                nums[j] = nums[j-1]  
                j -= 1
            # 直到找到一个比自己小的然后将 第 i 个插入
            nums[j] = temp

a = Solution()
array = [2,0,2,1,1,0]
a.sortColors(array)
print(array)
````

 # 学习一下评论区大佬的做法：
```
class Solution {
    public void sortColors(int[] nums) {
        int num0 = 0, num1 = 0, num2 = 0;
        for(int i = 0; i < nums.length; i++) {
            if(nums[i] == 0) {
                nums[num2++] = 2;
                nums[num1++] = 1;
                nums[num0++] = 0;
            }else if(nums[i] == 1) {
                nums[num2++] = 2;
                nums[num1++] = 1;
            }else {
                nums[num2++] = 2;
            }
        }
    }
}  
```

## 就是计算数列的边界吧，
- 当出现了一个2，那么就需要在数组最后面加上一个，
- 当出现一个1，那么就需要将之前的已有的所有2向后移动一位，同时在所有1的边缘处添加一个1
- 当出现一个0，那么就需要将之前的已有的所有1，2，向后移动一位同时在所有0 的边缘处添加一个0
## 很妙的是他对于0，1，2序列边缘的处理：
- 当出现2 那么只会影响到整体数组中2的上界，
- 同理出现了1，会影响1与2序列的上界，0同理
