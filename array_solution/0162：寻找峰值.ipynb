## 思路
- 先排除一头一尾
- 然后再从中间开始找


```python
from typing import List

class Solution:
    def findPeakElement(self, nums: List[int]) -> int:
        size = len(nums)
        if size == 1:
            return 0
        elif size >= 2:
            if nums[0] > nums[1]:
                return 0
            elif nums[size - 1] > nums[size - 2]:
                return size - 1

        # 将上面的特殊情况都排除了然后开始查找
        l, r = 1, size - 2
        while l < r:
            mid = (l + r) // 2
            if nums[mid] > nums[mid + 1] and nums[mid] > nums[mid - 1]:
                return mid
            elif nums[mid] < nums[mid + 1]:
                l = mid + 1
            elif nums[mid] < nums[mid - 1]:
                r = mid - 1
        if nums[l] > nums[l + 1] and nums[l] > nums[l - 1]:
            return l
        
a = Solution()
a.findPeakElement([1,2,1,3,5,6,4])
                  
```
