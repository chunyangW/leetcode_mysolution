- 思路
  - 双指针
    - 一共有k个数可以入选所以双指针最后间距要 == k - 1
    - 边界的判断就根据接近的定义来


```python
from typing import List

class Solution:
    def findClosestElements(self, arr: List[int], k: int, x: int) -> List[int]:
        size = len(arr)
        l, r = 0, size - 1
        while l < r and (r - l) > k - 1:
            if abs(arr[l] - x) > abs(arr[r] - x):
                l += 1
            elif abs(arr[l] - x) < abs(arr[r] - x):
                r -= 1
            else:
                r -= 1
        return arr[l : r + 1] # 切片不包含最后一个
```
