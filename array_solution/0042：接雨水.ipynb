- 

 height[left]  height[right]  leftMax  rightMax 

 height[left]<height[right] leftMax<rightMax()  left  leftMaxheight[left] left  left  1

 height[left]height[right] leftMaxrightMax right  rightMaxheight[right] right  right  1




```python
from typing import List

class Solution:
    def trap(self, height: List[int]) -> int:
        ans = 0
        left, right, leftMax, rightMax = 0, len(height) - 1, 0, 0
        while left < right:
            # 
            leftMax = max(leftMax, height[left])
            rightMax = max(rightMax, height[right])

            if height[left] <= height[right]:  #  (maxleftMax)
                ans += leftMax - height[left]
                left += 1
            else:
                ans += rightMax - height[right]
                right -= 1

        return ans
    
a = Solution()
a.trap([0,1,0,2,1,0,1,3,2,1,2,1])
```
