```python
# """
# This is MountainArray's API interface.
# You should not implement it, or speculate about its implementation
# """
class MountainArray:
    def __init__(self, Array):    #__init__是一个特殊的方法，称为类的构造函数。当你创建一个类的新实例时，
                                  #Python会自动调用这个方法。这个方法的目的是初始化新创建的对象的状态，即设置对象的初始属性值。
        self.array = Array

    def get(self, index: int) -> int:
        return self.array[index]

    def length(self) -> int:

        return len(self.array)
    

class Solution:
    def findInMountainArray(self, target: int, mountain_arr: 'MountainArray') -> None: 
        size = mountain_arr.length()
        l, r = 0, size - 1
        # 先找到山峰
        while l < r:
            mid = (l + r) // 2
            if mountain_arr.get(mid) < mountain_arr.get(mid + 1):
                l = mid
            elif mountain_arr.get(mid) < mountain_arr.get(mid - 1):
                r = mid
            else:
                peak = mid
                break
        # 然后分开找
        # 先左边
        l, r = 0, peak - 1
        while l < r:
            mid = (l + r) // 2
            if mountain_arr.get(mid) < target:
                l = mid + 1
            elif mountain_arr.get(mid) > target:
                r = mid - 1
            else:
                return mid
        if mountain_arr.get(l) == target:
            return l
        # 右边
        l , r = peak, size -1 
        while l < r:
            mid = (l + r) // 2
            if mountain_arr.get(mid) < target:
                r = mid - 1
            elif mountain_arr.get(mid) > target:
                l = mid + 1
            else:
                return mid
        if mountain_arr.get(l) == target:
            return l
        
        # 没找到
        return -1
         

array = MountainArray([1, 2, 3, 4, 5, 3, 1])

a = Solution()
a.findInMountainArray(3, array)
```
