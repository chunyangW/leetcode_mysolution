```python
# 二分查找的字符版本
# 但是不一样要找更大的
from typing import List

class Solution:
    def nextGreatestLetter(self, letters: List[str], target: str) -> str:
        l, r = 0, len(letters) - 1
        while l < r:
            mid = (l + r) // 2
            if (letters[mid] <= target):  # 只要小于等于那么久能往上走
                l = mid + 1
            elif (letters[mid] > target):
                r = mid  # 可能是答案
        if letters[l] > target:
            return letters[l]
        else:
            return letters[0]


a = Solution()
a.nextGreatestLetter(["c","f","j"], 'c')
```
