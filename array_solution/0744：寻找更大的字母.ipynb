```python
# 
# 
from typing import List

class Solution:
    def nextGreatestLetter(self, letters: List[str], target: str) -> str:
        l, r = 0, len(letters) - 1
        while l < r:
            mid = (l + r) // 2
            if (letters[mid] <= target):  # 
                l = mid + 1
            elif (letters[mid] > target):
                r = mid  # 
        if letters[l] > target:
            return letters[l]
        else:
            return letters[0]


a = Solution()
a.nextGreatestLetter(["c","f","j"], 'c')
```
