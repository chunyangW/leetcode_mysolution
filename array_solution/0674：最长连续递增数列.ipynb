```python
from typing import List
class Solution:
    def findLengthOfLCIS(self, nums: List[int]) -> int:
        res, temp = 1, 1
        size = len(nums)
        right = 0
        if size == 1:
            return 1
        while right < size - 1 :
            right += 1
            if nums[right] > nums[right - 1]:
                temp += 1
            else:
                res = max(temp, res)        
                temp = 1
        res = max(temp, res)
        return res    
            
```

```python
# 太巧妙了
class Solution:
    def findLengthOfLCIS(self, nums: List[int]) -> int:
        dp = [1] * len(nums)
        for i in range(1,len(nums)):
            if nums[i] > nums[i-1]:
                dp[i] = dp[i-1] +1
        return max(dp)
        
```
