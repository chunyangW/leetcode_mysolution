# 使用双指针来交换首尾的元素



```python
from typing import List

class Solution:
    def reverseString(self, s: List[str]) -> None:
        l, r = 0, len(s) - 1
        while l <= r:
            s[l], s[r] = s[r], s[l]
            l += 1
            r -= 1
        
a = Solution()
s =["H","a","n","n","a","h"]
a.reverseString(s)
print(s)
```
