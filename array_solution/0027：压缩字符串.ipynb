## 思路：
- 使用快慢指针， 慢指针指向最后要保留的元素的边界
- 快指针指向当前判断的元素

```python
from typing import List

class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        size = len(nums)
        fast, slow = 0, 0
        for fast in range(size):
            if nums[fast] == val:
                continue
            else:
                nums[slow] = nums[fast]
                slow += 1
        return slow 
```
