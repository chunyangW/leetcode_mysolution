```python
from typing import List

class Solution:
    def intersection(self, nums1: List[int], nums2: List[int]) -> List[int]:
        len1 = len(nums1)
        len2 = len(nums2)
        nums1.sort()
        nums2.sort()
        res = []
        i, j = 0, 0
        if len1 >= len2:
            while i< len1 and j < len2:
                if nums1[i] == nums2[j] and nums1[i] not in res :
                    res.append(nums1[i])
                    j += 1
                elif nums1[i] > nums2[j] :
                    j += 1
                else:
                    i += 1
        else:
            while j < len2 and i < len1:
                if nums1[i] == nums2[j] and nums1[i] not in res  :
                    res.append(nums1[i])
                    i += 1 # j是不断动了,i也得动    
                elif nums1[i] < nums2[j] :
                    i += 1
                else:
                    j += 1
                
        return res
```
