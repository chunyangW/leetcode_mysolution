```python
from typing import List

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        size = len(nums)
        if size == 1:
            return 1
        fast, slow = 0, 0
        while fast < size - 1:
            fast += 1
            if nums[fast] == nums[fast - 1]:
                continue
            else:
                nums[slow] = nums[fast - 1]
                slow += 1
        # 
        nums[slow] = nums[fast]
        slow += 1
        # print(nums[:slow])
        return slow    

a = Solution()
a.removeDuplicates([0,0,1,1,1,2,2,3,3,4])    
```
