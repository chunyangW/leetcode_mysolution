```python
from math import inf
from typing import List

class Solution:
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        # 其实是在找最大的窗口总数
        size = len(nums)
        Sum, res = 0, -inf  # 一定要注意其实会有负数的答案， 如果初始值是0那么最终结果肯定不是0
        left, right = 0, 0
        while right < size :
            # 每次进循环首先将窗口填满
            while right - left <= k - 1:
                Sum += nums[right]
                right += 1
            # 然后判断该窗口总数是不是更大
            res = max(res, Sum)
            # 最后移动窗口并将最左边的元素弹出
            Sum -= nums[left]
            left += 1
        return res / k 
a = Solution()
a.findMaxAverage([1,12,-5,-6,50,3], 4)
```
