```python
from math import inf
from typing import List

class Solution:
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        # 
        size = len(nums)
        Sum, res = 0, -inf  #  00
        left, right = 0, 0
        while right < size :
            # 
            while right - left <= k - 1:
                Sum += nums[right]
                right += 1
            # 
            res = max(res, Sum)
            # 
            Sum -= nums[left]
            left += 1
        return res / k 
a = Solution()
a.findMaxAverage([1,12,-5,-6,50,3], 4)
```
