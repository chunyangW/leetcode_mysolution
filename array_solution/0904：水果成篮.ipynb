```python
from typing import List
from collections import defaultdict

class Solution:
    def totalFruit(self, fruits: List[int]) -> int:
        # 找到最多包含两种元素的最长数组
        cnt = defaultdict(int)
        res = 0
        k = 2  # 最多两种元素 
        left = 0
        for right, val in enumerate(fruits):
            cnt[val] += 1
            while len(cnt) > k:
                if cnt[fruits[left]] == 1:
                    del cnt[fruits[left]]
                else:
                    cnt[fruits[left]] -= 1
            res = max(res, right - left + 1)
        return res
a = Solution()
a.totalFruit([0,1,2,2])
```
