```python
from typing import List

class Solution:
    def longestSubarray(self, nums: List[int]) -> int:
        # 存在的0 数目不能大于 1
        size = len(nums)
        right, left, res, temp, count = 0, 0, 0, 0, 0
        while right < size:
            if nums[right] == 1:
                temp += 1
            else:
                count += 1
            right += 1

            while count > 1:
                temp -= 1
                if nums[left] == 0:
                    count -= 1
                    temp += 1
                left += 1
                
            res = max(res, temp)
        return res if count != 0 else res - 1

a = Solution()
a.longestSubarray([1,1,1])
```
