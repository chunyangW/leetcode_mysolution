- 要考虑本来就是有重复的字符串
  - lee 与 llee
- 就是找交集是不是name!!

```python
from typing import List

class Solution:
    def isLongPressedName(self, name: str, typed: str) -> bool:
        len1 = len(name)
        len2 = len(typed)
        index1, index2 = 0, 0

        while index1 < len1 and index2 < len2:
            s1 = name[index1]
            s2 = typed[index2]
            if s1 == s2:
                index1 += 1
                index2 +=1
            elif index1 > 0 and s2 == name[index1 - 1]:
                index2 += 1 # typed落后了要继续加
            else:
                return False
          #结尾  a  a 同时结束   ex exa name 先结束但是错误，gn gnn name先结束但是正确 所以需要判断
        if index1 == len1 and index2 == len2:# 同时结束型
            return True
        if index1 == len1 :# 判断len2 剩下的是不是都等于name最后
            while index2 < len2 and typed[index1] == name[-1]:
                index2 += 1
            if index2 == len2:
                return True
            else:
                return False
        else:
            return False

 
    
a = Solution()
print(a.isLongPressedName("alex", "aaleexa"))
```

```python
class Solution:
    def isLongPressedName(self, name: str, typed: str) -> bool:
        i = 0
        name += "1"
        for s in typed:
            if s == name[i]:
                i += 1
            elif s != name[i - 1]:
                return False
        return i == len(name) - 1

a = Solution()
print(a.isLongPressedName("alex", "aaleexa"))
```
