### 思路
先找增加到极大值然后下降的
在下降到最后才返回
不然都返回0

```python
from typing import List

class Solution:
    def longestMountain(self, arr: List[int]) -> int:
        size = len(arr)
        fast, slow, flag = 0, 0, 0
        ans = 0
        # falg = 0 是初始上升， flag = 1是上升期， flag = 2是下降期
        for fast in range(1, size):
            if arr[fast] > arr[fast - 1] and (flag == 0 or flag == 1):
                flag = 1 # 有成为山脉的潜力
            elif arr[fast] < arr[fast - 1] and (flag == 1 or flag == 2):
                flag = 2 # 已经是山脉了 同时能够排除掉上面的判断
            else:
                if flag == 2: # 山脉断了
                    ans = max(ans, fast - slow + 1 )
                slow = fast
                flag = 0
        return ans

a = Solution()
a.longestMountain([0,1,2,3,4,5,4,3,2,1,0])
```
