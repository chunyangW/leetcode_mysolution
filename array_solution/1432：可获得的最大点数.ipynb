```python
from typing import List
from math import inf
class Solution:
    def maxScore(self, cardPoints: List[int], k: int) -> int:
        # 其实实际上是在找中间总和是最小的窗口
        size = len(cardPoints)
        left, right, temp = 0, 0, 0
        res = sum(cardPoints)
        window_len = size - k 
        while right < size and left <= right:
            while right - left <= window_len - 1:
                temp += cardPoints[right]
                right += 1
            res = min(res, temp)
            temp -= cardPoints[left]
            left += 1
        return sum(cardPoints) - res
        
a = Solution()
a.maxScore([9,7,7,9,7,7,9], 7)
```
