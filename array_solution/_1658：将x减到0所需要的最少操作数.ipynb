```python
from typing import List

class Solution:
    def minOperations(self, nums: List[int], x: int) -> int:
        # 
        # 
        Sum = sum(nums)
        target = Sum - x   #
        size = len(nums)
        left, right, window, MAXlen = 0, 0, 0, -1
        # 
        while right < size: 
            if right < size : # 
                window += nums[right]
                right += 1
            while window > target and left < right:
                window -= nums[left]
                left += 1
            if window == target:
                MAXlen = max(right - left, MAXlen)
        return size - MAXlen  if MAXlen != -1 else -1                                                          
    
a = Solution()
a.minOperations([1,1,4,2,3],5)
```
