# 要判断是不是元音字符
- 不是就往下面继续找
- 找到才交换
# 将字符串转化为可迭代的对象
- s = str(s)
# 将列表字符串重新变成字符串
- "".join(s)  ""中的内容表示连接相邻两个元素时中间需要添加的东西

```python
class Solution:
    def reverseVowels(self, s: str) -> str:
        # 先写一个用于判断是否是元音字符的子函数
        def isTrue(ch : str) -> bool:
            return ch in "aeiouAEIOU"
        
        # 主函数
        l, r = 0, len(s) - 1
        s = list(s)
        while l <= r:
            while l <= r and not isTrue(s[l]):
                l += 1
            while l <= r and not isTrue(s[r]):
                r -= 1
            if l <= r:
                s[l], s[r] = s[r], s[l]
                l += 1
                r -= 1

        return "".join(s)  


a = Solution()
string = "leetcode"
print(a.reverseVowels(string))
```
