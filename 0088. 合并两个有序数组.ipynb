## On+mn

```python
from typing import List

class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        for i in range(n):   # n 
            j = i
            temp = nums2[i]
            while temp < nums1[m - 1 + j]:        # m - 1 + i 
                nums1[m + j] = nums1[m - 1 + j]# 
                j -= 1
            nums1[m + j] = temp
        print(nums1)
        
a = Solution()
a.merge([2,0], 1, [1], 1)
```
