# 思路
- 和插入排序类似啊，就是相当于num1为左边的排序数组 然后 Num2为未排序数组然后一个个比较插入就好了
## 但是怎么实现O（n+m）n

```python
from typing import List

class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        for i in range(n):   # 总共要插入n 个元素
            j = i
            temp = nums2[i]
            while temp < nums1[m - 1 + j]:        # m - 1 + i 是未排序的边缘
                nums1[m + j] = nums1[m - 1 + j]# 将更大的数字向右移动
                j -= 1
            nums1[m + j] = temp
        print(nums1)
        
a = Solution()
a.merge([2,0], 1, [1], 1)
```
