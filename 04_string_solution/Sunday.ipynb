- 他是从左向右查找
- 如果出现了不相同的
  - 先看待查找串中模式串的右边一个
    - 如果都没有出现在模式串中，那么直接平移 模式串长度跳过它
    - 如果出现过，那么尽可能平移小一点，平移到最右边出现过的和它对齐，也就是 m - i
![屏幕截图 2024-10-19 180721.png](<attachment:屏幕截图 2024-10-19 180721.png>)

```python
class Solution:
    def Sunday(self, p: str, t: str)->int:
        # 先生成字典来查找最右边出现过的元素
        bcTable = {}
        m = len(p)
        n = len(t)
        # 从左向右遍历会将前面相同的覆盖
        for i in range(m):
            bcTable[p[i]] = m - i # 需要移动多少
        
        
        i = 0
        while i <= n - m:
            # while t[i + j] == p[j]:
            #     j += 1
            # 可以直接切片
            if t[i:i+m] == p:
                return i
            # 排除 i+m = n
            if i + m == n:
                return -1
            i += bcTable.get(t[i + m], m + 1)
        return -1 # 没找到

test = Solution()
test.Sunday("bba", "aaaaa")     
            
```
