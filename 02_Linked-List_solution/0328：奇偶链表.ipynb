```python
from typing import Optional
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def oddEvenList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        dummy_node1 = ListNode(0, head)
        dummy_node2 = ListNode()
        dummy_node2.next = head.next
        cur1 = dummy_node1.next 
        cur2 = dummy_node2.next
        flag = 1
        while cur1 != None and cur2 != None:
            if flag:
                cur1.next = cur2.next
                if cur1.next == None:  # 只要奇数做个存档
                    temp = cur1 
                cur1 = cur1.next
                flag = 0
                
            else:
                cur2.next = cur1.next                
                cur2 = cur2.next
                flag = 1
        # flag 还有一个好处是知道是谁最后连着none    1 是偶数后面连none
        if flag:
            cur1.next = dummy_node2.next
        else:
            temp.next = dummy_node2.next
        return dummy_node1.next
        



array = [1, 5, 2, 6, 3, 7, 4]
head = ListNode(array[0])
cur = head
for i in range(1, len(array)):
    node = ListNode(array[i])
    cur.next = node
    cur = cur.next

test = Solution()
head = test.oddEvenList(head)

cur = head
while cur != None:
    print(cur.val, end="->")
    cur = cur.next
print("None")




```
