```python
class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        f = head
        s = head

        while f and f.next:
            f = f.next.next
            s = s.next
            if s == f:
                s = head
                while s != f:
                    s = s.next
                    f = f.next
                return s
        return None
        
```
