- 假设 headA 长度为 a + c（a 是独立部分长度，c 是相交部分长度），headB 长度为 b + c（b 是独立部分长度，c 是相交部分长度）。经过两个指针 分别遍历两个链表之后，两个指针都走了 a + b + c 的距离，因此能够保证在交点处相遇。

```python
class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        # 跑全程法，当A跑完从B头开始跑，同理B也是这么跑，那么一定在两个相交的地方相交
        A, B = headA, headB
        while A != B:
            if A == B:
                return A
            A = A.next if A else headB  # 三元赋值语句，如果A当前有值，那么就下一个如果A为none，那么就去B头
            B = B.next if B else headA
        return A
```

```python

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        # 竟然可以头节点就可以开始相交
        if not headA or not headB :
            return None 
        # 先把两个链表长度求出来，让长的移动到和短的相等再判断1
        sizeA, sizeB = 1, 1
        curB = headB
        curA = headA
        while curA.next :
            curA = curA.next
            sizeA += 1 
        while curB.next:
            curB = curB.next
            sizeB += 1

        curA = headA
        curB = headB
        if sizeB >= sizeA:
            count = sizeB - sizeA
            for _ in range(count):
                curB = curB.next
        else:
            count = sizeA - sizeB
            for _ in range(count):
                curA = curA.next
        
        while curA:# 肯定不是头就相交了，不，可以
            if curA == curB:
                return curA
            curA = curA.next
            curB = curB.next
        return None
```
