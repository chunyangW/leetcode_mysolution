```python
class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        # ABB
        A, B = headA, headB
        while A != B:
            if A == B:
                return A
            A = A.next if A else headB  # AAnoneB
            B = B.next if B else headA
        return A
```
