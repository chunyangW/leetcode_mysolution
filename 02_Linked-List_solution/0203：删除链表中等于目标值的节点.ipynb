```python
# Definition for singly-linked list.

from typing import Optional
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:
        # 因为有可能删除掉头节点，而删除节点最重要的是需要知道前面的一个节点
        # 那么就会出现分类讨论等情况
        # 使用dummy node 节点就能使得代码更加简洁
        dummy_node = ListNode()
        dummy_node.next = head
        cur = dummy_node # cur表示要处理的上面一个节点因为删除必须知道要删除的上面一个节点
        while(cur.next != None):
            if(cur.next.val == val):# 如果相等那么下一个节点被删除
                cur.next = cur.next.next
            else:  # 否则下一个节点一定被保留，所以可以安全移动到下一个去
                cur = cur.next
        return dummy_node.next
    

# 创建原始链表
array = [1, 1, 1, 2, 1, 2, 3]
head = ListNode(array[0])   # 头要单独保存起来
cur = head                  # 然后用一个指针来不断生成下一连接
for i in range(1, len(array)):
    node = ListNode(array[i])
    cur.next = node   # 连接
    cur = cur.next 

test = Solution()
head = test.removeElements(head, 1)   #返回的head要修改原来的head不然等于没删除

# 显示当前链表
cur = head
while(cur):
    print(cur.val, end = "->")
    cur = cur.next
print("None")
```
