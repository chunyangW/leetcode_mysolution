```python
class Linked_node:  # 使用一个类来作为创建链节点可以附带多个链表所包含的属性
    def __init__(self, data=0, next=None):  # 使用默认参数
        self.val = data
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None

    def Linked_List_Creat(self, data):
        if not data:
            return
        self.head = Linked_node(data[0])
        cur = self.head
        for i in range(1, len(data)):           # 重复创建新的节点的过程
            node = Linked_node(data[i])
            cur.next = node
            cur = cur.next
    
    def insert_Head(self, data = 0):  # 先创建一个新的节点   然后将新的节点的next指向原来的头   最后将head赋值为新的头了   
        node = Linked_node(data)
        node.next = self.head
        self.head = node
    
    def insert_Rear(self, data = 0):
        # 先创建新节点，然后不断后移动直到下一个是 None
        node = Linked_node(data)
        cur = self.head
        while cur.next:  # 一定是下一个不是None 才能保证跳出为最后一个
            cur = cur.next
        cur.next = node

    def insert_before_inside(self, data = 0, index = 1):  # 插入到序号为1 的元素之前也就是在0之后 那么插入之后，插入的位置索引刚好是 index(从0开始)
        cur = self.head
        count = 0
        while cur and count < index - 1:
            count += 1
            cur = cur.next

        if not cur:
            return "Error"
        
        node = Linked_node(data)
        node.next = cur.next  # 很重要，先将新插入的与cur后面的一个连接上
        cur.next = node

    def change_data(self, data = 0, index = 0): # 改变元素
        cur = self.head
        count = 0
        while cur and count < index :
            count += 1
            cur = cur.next
        if not cur:
            raise TypeError("indexError")
        cur.val = data

    # 删除链表头部，尾部，中间元素
    def del_head(self):
        if self.head:
            self.head = self.head.next  # head赋值为之前 head 的下一个
    
    def del_inside(self, index = 1):
        cur = self.head
        count = 0
        while cur and count < index - 1: # while 会停止在取等，同时count与索引相同
            count += 1
            cur = cur.next
        if not cur:
            return None
        cur.next = cur.next.next  # 连用两个超过中间的
         
    def del_rear(self):
        if not self.head or not self.head.next:
            return "Error"

        cur = self.head
        while cur.next.next: # 终止在cur.next.next = None
            cur = cur.next
        cur.next = None
    
# 使用 LinkedList 类创建链表
test = LinkedList()
test.Linked_List_Creat([1, 2, 3, 4, 5])
test.insert_Head(66)
test.insert_Rear(6)
test.insert_before_inside(9, 2)
test.change_data(100, 0)
test.del_inside(1)
test.del_rear()

# 打印链表以验证
current = test.head
while current:
    print(current.val, end=" -> ")
    current = current.next
print("None")
```
