```python
from typing import Optional


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # 用来指示一个变量可以是None，类型也可以是其他的选项
        cur = head
        while(cur != None):
            # 如果 cur 是空节点None那么就可以提前终止了，但是下一个是空节点不能提前终止，因为最后一个也可能删掉
            while(cur.next != None and cur.next.val == cur.val):  # 一定要保证cur.next不是None 才能有 cur.next.val 因为None没有.val属性
                cur.next = cur.next.next
            cur = cur.next # 从上面出来后一定前面是只出现一次的元素
        return head



# 创建原始链表
array = [1, 1, 1, 2, 2, 3]
head = ListNode(array[0])   # 头要单独保存起来
cur = head                  # 然后用一个指针来不断生成下一连接
for i in range(1, len(array)):
    node = ListNode(array[i])
    cur.next = node   # 连接
    cur = cur.next 

test = Solution()
head = test.deleteDuplicates(head)

# 显示当前链表
cur = head
while(cur):
    print(cur.val, end = "->")
    cur = cur.next
print("None")
```
