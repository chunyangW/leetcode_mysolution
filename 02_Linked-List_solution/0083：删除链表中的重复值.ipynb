```python
from typing import Optional


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # None
        cur = head
        while(cur != None):
            #  cur None
            while(cur.next != None and cur.next.val == cur.val):  # cur.nextNone  cur.next.val None.val
                cur.next = cur.next.next
            cur = cur.next # 
        return head



# 
array = [1, 1, 1, 2, 2, 3]
head = ListNode(array[0])   # 
cur = head                  # 
for i in range(1, len(array)):
    node = ListNode(array[i])
    cur.next = node   # 
    cur = cur.next 

test = Solution()
head = test.deleteDuplicates(head)

# 
cur = head
while(cur):
    print(cur.val, end = "->")
    cur = cur.next
print("None")
```
