## 这是一个不定长度移动窗口
最关键的是要统计是不是重复了
- 用哈希表统计
- 如果有重复了怎么办
- 先统计长度更新最长
- 左边直接移动到重复的位置

```python
from typing import List

class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        window = {}  # 哈希表窗口
        size = len(s)
        ans = 0
        left = 0
        for right in range(size):
            if s[right] not in window:
                window[s[right]] = 1
            else:
                window[s[right]] += 1

            while window[s[right]] > 1: # 如果当前出现重复了可能是left 到 right 之间某一处出现了问题所以需要不断前进left
                window[s[left]] -= 1
                left += 1
                
            ans = max(ans, right - left + 1 )
        return ans
    

    
a = Solution()
a.lengthOfLongestSubstring("dvdf")
```
