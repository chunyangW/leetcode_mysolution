```python
class Solution:
    def equalSubstring(self, s: str, t: str, maxCost: int) -> int:
        # 还是一样的套路，右边每次移动，然后左边在满足条件下收缩
        size = len(s)
        right, left, temp, res = 0, 0, 0, 0
        while right < size:
            temp += abs(ord(s[right]) - ord(t[right]))  # ord 函数用来获得字符的ascii码值

            while temp > maxCost:
                temp -= abs(ord(s[left]) - ord(t[left]))
                left += 1
        
            res = max(res, right - left + 1)
            right += 1
        return res
    
a = Solution()
a.equalSubstring("pxezla", "loewbi", 25)
```
