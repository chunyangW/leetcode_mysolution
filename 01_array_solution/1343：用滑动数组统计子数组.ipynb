```python
from typing import List

arr = [1,2,3,4]
def popleft(window: List[int]):
    window[: -1] = window[1: ]
    window.pop()
    print(window)

popleft(arr)
```

```python
from typing import List

class Solution:
    def numOfSubarrays(self, arr: list[int], k: int, threshold: int) -> int:
        left, right = 0, 0
        window = 0
        count = 0

        while right < len(arr):  
            window += arr[right]  # 为什么要先加而不是后加这一行放最后就错了 后加就要在下一个循环中判断了所以关键
            # 而且一开始不是right = left = 0是吧，说明window有arr[0]但是如果放在后面就等于没有这个元素判断的时候
            if  right - left + 1 >= k: # 窗口等于k了 这两个条件不一定同时满足 那窗口相等了一定要满足left++啊
                if window >= k * threshold:
                    count += 1
                window -= arr[left]
                left += 1
            right += 1
        return count


a = Solution()
a.numOfSubarrays([2,2,2,2,5,5,5,8],3,4)
```
