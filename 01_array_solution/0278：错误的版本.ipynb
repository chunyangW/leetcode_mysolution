```python
# The isBadVersion API is already defined for you.
def isBadVersion(version: int) -> bool:
    if version >=4 :
        return True
    else:
        return False
    

class Solution:
    def firstBadVersion(self, n: int) -> int:
        l, r = 1, n
        while l < r:
            mid = (l + r) // 2
            if isBadVersion(mid): # 说明错误的版本还在左边
                r = mid 
            else:
                l = mid + 1
        return l
    
        
```
