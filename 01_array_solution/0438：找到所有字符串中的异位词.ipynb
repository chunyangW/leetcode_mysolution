```python
from typing import List
from collections import Counter
class Solution:
    def findAnagrams(self, s: str, p : str) -> List[int]:
        # 使用哈希表来比较元素是不是相等的
        state1 = Counter(l for l in p)
        def check (state1, state2):
            for key in state1:
                if state1[key] != state2[key]:
                    return False
            return True
        k = len(s)
        size = len(p)
        state2 = Counter()
        res = []
        if k < len(p):
            return []
        for right in range(size):
            state2[s[right]] += 1
        if check(state1, state2):
            res.append(right - size + 1)
        for right in range(size, k):
            state2[s[right]] += 1
            state2[s[right - size]] -= 1
            if check(state1, state2):
                res.append(right - size + 1)
        return res
    
a = Solution()
a.findAnagrams("cbaebabacd", "abc")
        
```
