# 双指针同时最重要的是要能够将非字母数字字符移除

```
import re

def remove_non_alphanumeric(text):
    # 使用正则表达式替换掉非字母数字的字符
    return re.sub(r'[^a-zA-Z0-9]', '', text)    # re.sub(substitute 替换)

# 示例使用
input_text = "Hello, World! 123."
cleaned_text = remove_non_alphanumeric(input_text)
print(cleaned_text)  # 输出: HelloWorld12

```

````python
import re


class Solution:
    def isPalindrome(self, s: str) -> bool:
        # 先将非字母数字排除
        def removeNonAlphaNum(s):
            return re.sub(r'[^a-zA-Z0-9]', '', s)
        # 再使用双指针判断是不是回文序列
        s = removeNonAlphaNum(s)
        l, r = 0, len(s) - 1
        #s = list(s) #去字符串可以迭代并引用标签
        # 别忘了统一大小写

        while l <= r:
            if s[l].lower() != s[r].lower():
                return False
            l += 1
            r -= 1
        return True
    
a = Solution()
a.isPalindrome("A man, a plan, a canal: Panama")
````
