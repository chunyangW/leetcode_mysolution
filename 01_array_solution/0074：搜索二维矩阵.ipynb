```python
from typing import List


class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        sizei = len(matrix)   # 
        sizej = len(matrix[0])# 

        # 
        def indexconvert(index):
            i = index // sizej
            j = index % sizej   #j = index - sizej *(i)
            return matrix[i][j]
        
        l, r = 0, sizei * sizej - 1
        while l < r:
            mid = (l + r) // 2
            if indexconvert(mid) < target:
                l = mid + 1 
            elif indexconvert(mid)  > target:
                r = mid - 1
            else:
                return True
        if indexconvert(l) == target:
            return True
        else:
            return False

a = Solution()
a.searchMatrix([[1,3,5,7],[10,11,16,20],[23,30,34,60]],60)
```
