```python
from typing import List

class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        
        l, r = 0, len(nums) - 1
        # 使用小于号那么最后跳出一定是在l处
        while l < r: 
            mid = (l + r) // 2
            if nums[mid] < target:
                l = mid + 1
            elif nums[mid] > target:
                r = mid -1
            elif nums[mid] == target:
                return  mid
        return l if nums[l] >= target else l + 1
    

a = Solution()
a.searchInsert([1,3,5,6], 0)
```
