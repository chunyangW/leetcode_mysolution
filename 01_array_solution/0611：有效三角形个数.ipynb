## 得有一个子函数用来判断是不是三角形
- 构成三角形的条件为：任意两边和大于第三边，或者任意两边差小于第三边。只要满足这两个条件之一就可以构成三角形。
- 那么就是说如果我们将三条边升序排序，假设 $a \le b \le c$，则如果满足 $a + b > c$，则 $a + c > b$ 和 $b + c > a$ 一定成立。

# 将所有的元素升序排列然后将第三条边从 $i = 2 $开始 
# 然后左边两个对撞指针来找剩下两个边
- 因为是要左边两个相加大于第三个边就好，那么
- 当相加小于等于的时候说明第一条边太短了（第二条边也不能变小，变大的已经统计过了），要left + 1
- 当相加大于的时候说明第一条边可以取从left 到 right - 1的所有值 此时right - 1

```python
from typing import List

class Solution:
    def triangleNumber(self, nums: List[int]) -> int:
        nums = sorted(nums)
        size = len(nums)
        i, ans = 2, 0
        while i < size:
            left, right = 0, i - 1
            while left < right:
                if nums[left] + nums[right] <= nums[i]:
                    left += 1
                else:
                    ans += (right - left)
                    right -= 1
            i += 1
        return ans
a = Solution()
a.triangleNumber([4,2,3,4])
            
```
