```python
from typing import List

class Solution:
    def numRescueBoats(self, people: List[int], limit: int) -> int:
        people = sorted(people)
        size = len(people)
        l, r = 0, size - 1
        ans = 0
        while l < r:
            if people[l] + people[r] > limit:
                ans += 1
                r -= 1
            else:
                ans += 1
                l += 1
                r -= 1
        if l == r: # 
            ans += 1
        return ans
    
```
