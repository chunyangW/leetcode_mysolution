## 思路
- 排序后将一个放大的一共放小的看看能不能塞进去？
- 因为它最多只能坐两个人
- 所以最优化的是大的配小的

```python
from typing import List

class Solution:
    def numRescueBoats(self, people: List[int], limit: int) -> int:
        people = sorted(people)
        size = len(people)
        l, r = 0, size - 1
        ans = 0
        while l < r:
            if people[l] + people[r] > limit:
                ans += 1
                r -= 1
            else:
                ans += 1
                l += 1
                r -= 1
        if l == r: # 最后还剩一个
            ans += 1
        return ans
    
```
