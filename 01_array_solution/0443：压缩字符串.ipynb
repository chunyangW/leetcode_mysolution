## 跑长码（游程编码）


```python
# 快慢指针
# 快指针一直走慢指针停在开头 直到快指针找到不一样的慢指针重置位置
from typing import List

class Solution:
    def compress(self, chars: List[str]):
        f, l = 0, 0
        ans = []
        size = len(chars)
        for f in range(size):
            if chars[f] == chars[l]:
                continue
            else:
                # 长度为1
                if f - l == 0:
                    ans.append(chars[l])
                    l = f
                # 长度不为一
                else:
                    ans.append(chars[l])
                    temp = f - l
                    l = f
                    # 有个问题怎么拆开数字为字符呢
                    # 先变成字符串然后一个个添加
                    # 然后由于循环中是一个个字符循环的
                    temp = str(temp)
                    for x in temp:
                        ans.append(x)
        # 不要忘记最后一串
        ans.append(chars[l])
        temp = str(f - l + 1)
        for x in temp:
            ans.append(x)
            
        return ans
    
a = Solution()
a.compress(["a","a","b","b","c","c","c"])
                    
```

```python
class Solution:
    
    def compress(self, chars: List[str]) -> int:
        def reverse(left, right):
            while left < right:
                chars[left], chars[right] = chars[right], chars[left]
                left += 1
                right -= 1

        slow, fast = 0, 0
        while fast < len(chars):
            fast_start = fast
            while fast + 1 < len(chars) and chars[fast + 1] == chars[fast]:
                fast += 1
            
            chars[slow] = chars[fast_start]
            slow += 1

            if fast - fast_start > 0:
                cnt = fast - fast_start + 1
                slow_start = slow
                while cnt != 0:
                    chars[slow] = str(cnt % 10)
                    slow += 1
                    cnt = cnt // 10
                reverse(slow_start, slow - 1)
            
            fast += 1
        return slow
    
a = Solution()
a.compress(["a","a","b","b","c","c","c"])
```
