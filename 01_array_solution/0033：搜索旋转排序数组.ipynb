```python
from typing import List

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        l, r = 0, len(nums) - 1
        while l < r:
            mid = (l + r) // 2
            if nums[mid] > nums[r]:  # 
                l = mid +  1
            elif nums[mid] < nums[l]: # mid
                r = mid
            else:                      # mid
                r -= 1             # rr--
        # l 
        mi = l 
        
        # 
        l, r = 0, len(nums) - 1
        if nums[0] <= target <= nums[mi]:
            r = mi
        else:
            l = mi
        while l < r:
            mid = (l + r) // 2
            if nums[mid] < target :
                l = mid + 1
            elif nums[mid] > target:
                r = mid - 1
            else:
                return mid
        return l
```
