```python
from typing import List
class Solution:
    def maxSatisfied(self, customers: List[int], grumpy: List[int], minutes: int) -> int:
        # 
        Sum = 0
        size = len(customers)
        for i in range(size):
            if  not grumpy[i]:
                Sum += customers[i]
        # 
        left , right = 0, 0
        temp, Convert = 0, 0
        while right < size:
            # 
            while right - left <= minutes - 1:
                if grumpy[right]:
                    temp += customers[right]
                right += 1
            Convert = max(temp, Convert)
            # 
            if grumpy[left]:
                temp -= customers[left]
            left += 1
        return Convert + Sum
```
