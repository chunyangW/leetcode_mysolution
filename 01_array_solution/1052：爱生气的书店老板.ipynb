```python
from typing import List
class Solution:
    def maxSatisfied(self, customers: List[int], grumpy: List[int], minutes: int) -> int:
        # 先求出不使用控制技巧时满意的顾客数目
        Sum = 0
        size = len(customers)
        for i in range(size):
            if  not grumpy[i]:
                Sum += customers[i]
        # 然后再滑动窗口求出使用技巧而能转变满意态度的最大数目
        left , right = 0, 0
        temp, Convert = 0, 0
        while right < size:
            # 每次进循环首先将窗口填满
            while right - left <= minutes - 1:
                if grumpy[right]:
                    temp += customers[right]
                right += 1
            Convert = max(temp, Convert)
            # 再将最左边的减掉
            if grumpy[left]:
                temp -= customers[left]
            left += 1
        return Convert + Sum
```
