```python
from typing import List

class Solution:
    def minOperations(self, nums: List[int], x: int) -> int:
        # 只能移动最左边与最右边、
        # 相当于找一个最长的窗口然后外面的元素等于目标值
        Sum = sum(nums)
        target = Sum - x   #窗口元素需要的总和
        size = len(nums)
        left, right, window, MAXlen = 0, 0, 0, -1
        # 使用滑动窗口，相当于遍历所有能相等的窗口
        while right < size: 
            if right < size : # 滑动窗口就是每次进来大循环就要向右移动，不受其他影响
                window += nums[right]
                right += 1
            while window > target and left < right:
                window -= nums[left]
                left += 1
            if window == target:
                MAXlen = max(right - left, MAXlen)
        return size - MAXlen  if MAXlen != -1 else -1                                                          
    
a = Solution()
a.minOperations([1,1,4,2,3],5)
```
